```python
import pandas as pd
import numpy as np
import folium
import matplotlib.pyplot as plt
import gmaps
```

```python
alfa = pd.read_csv('alfa_yogya.csv')
indo = pd.read_csv('indo_yogya.csv')
```

```python
map = folium.Map(location=[-7.77931364798441, 110.36706875152824],
                 zoom_start=10, control_scale=True, tiles='Stamen Toner')

#alfamart
for lat, lng in zip(alfa['Lat'],alfa['Lng']):
    folium.Marker(location=[lat,lng], icon=folium.Icon(color='red', icon='glyphicon-shopping-cart')).add_to(map)

#indomaret
for lat, lng in zip(indo['Lat'],indo['Lng']):
    folium.Marker(location=[lat,lng], icon=folium.Icon(color='blue', icon='glyphicon-shopping-cart')).add_to(map)

#circlemap indomaret radius 20 meter
for lat, lng in zip(indo['Lat'],indo['Lng']):
    folium.Circle(location=[lat,lng], radius=100).add_to(map)

#circlemap alfa 20 meter
for lat, lng in zip(alfa['Lat'],alfa['Lng']):
    folium.Circle(location=[lat,lng], radius=100, color='red').add_to(map)

```
