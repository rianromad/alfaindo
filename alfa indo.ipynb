```python
def scrape(banks):
    lat_ = []
    lng_ = []
    name_ = []
    address_ = []
    rating_ = []
    user_ratings_total_ = []

    #ambil data dari json
    for bank in banks:
        for data in bank:
            name = data['name']
            location = data['geometry']['location']
            address = data['formatted_address']
            rating = data['rating']
            user_ratings_total = data['user_ratings_total']

            name_.append(name)
            lat_.append(location['lat'])
            lng_.append(location['lng'])
            address_.append(address)
            rating_.append(rating)
            user_ratings_total_.append(user_ratings_total)
        
    return lat_, lng_, name_, address_, rating_, user_ratings_total_
```
